```python
import torch
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA/Hardware Acceleration Available: {torch.cuda.is_available()}")

# 1. Simulate a wireless complex signal using NumPy
fs = 1000  # Sampling frequency
t = np.arange(0, 1, 1/fs)
# Complex exponential representing a baseband signal (I + jQ)
complex_signal = np.exp(1j * 2 * np.pi * 50 * t) 

# 2. Convert it to a PyTorch Complex Tensor
tensor_signal = torch.tensor(complex_signal, dtype=torch.complex64)

# 3. Quick plot of the Real (I) and Imaginary (Q) components
plt.figure(figsize=(10, 4))
plt.plot(t[:100], tensor_signal.real[:100], label="In-Phase (I / Real)")
plt.plot(t[:100], tensor_signal.imag[:100], label="Quadrature (Q / Imag)")
plt.title("Verified: PyTorch & NumPy Handling Complex Wireless Signals")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
